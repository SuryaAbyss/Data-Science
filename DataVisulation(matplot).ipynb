# Labels

```python
x = [3,8,1,0]
y = [6,2,7,11]
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x = [3,8,1,0]
y = [6,2,7,11]
plt.plot(x)
plt.plot(y)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

# Title

```python
plt.plot(x)
plt.plot(y)
plt.title('Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
modification = {'color':'red','size':15 ,"family":'serif'}
plt.title('Height vs Weight',fontdict=modification)
plt.xlabel('Height')
plt.ylabel('Weight',fontdict=modification)
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
modification = {'color':'red','size':15 ,"family":'serif'}
plt.title('Height vs Weight',fontdict=modification,loc='left') #loc=location
plt.xlabel('Height')
plt.ylabel('Weight',fontdict=modification)
plt.show()
```

# Grid

```python
plt.plot(x)
plt.plot(y)
plt.title('Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid()
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.title('Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid(axis='x')  #axis=x or y it can also be customized
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.title('Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid(ls = '--',lw=3)  #axis=x or y it can also be customized
plt.show()
```

# Ticks and Tick lebel

```python
x = [3,8,1,0]
y = [6,2,7,11]

plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,2,2])
```

```python
x = [3,8,1,0]
y = [6,2,7,11]

fig = plt.figure()
ax = fig.add_axes([1,1,2,2])

ax.plot (x, y)
ax.set_title("Animal")
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

ax.set_xticklabels([100,102,103,104]) # Fixed typo here

plt.plot(x,y)
plt.show()
```
