```python
count = 15
Player = []
Span = []
Mat = []
Inns = []
NO = []
Runs = []
HS = []
Ave = []
BF = []
SR = []
Hundreds = []
FiftyS = []
Zeroes = []
Fours = []
Sixes = []

for i in soup.findAll("td"):
    # Update the pattern in re.sub to extract content properly
    i = re.sub(r'<.*?>|</.*?>', '', str(i))
    if count == 15:
        Player.append(i)
    elif count == 0:
        Span.append(i)
    elif count == 1:
        Mat.append(i)
    elif count == 2:
        Inns.append(i)
    elif count == 3:
        NO.append(i)
    elif count == 4:
        Runs.append(i)
    elif count == 5:
        HS.append(i)
    elif count == 6:
        Ave.append(i)
    elif count == 7:
        BF.append(i)
    elif count == 8:
        SR.append(i)
    elif count == 9:
        Hundreds.append(i)
    elif count == 10:
        FiftyS.append(i)
    elif count == 11:
        Zeroes.append(i)
    elif count == 12:
        Fours.append(i)
    elif count == 13:
        Sixes.append(i)

    count = (count + 1) % 16  # Reset count to 0 after reaching 15

# Print or use the extracted data as needed
```
